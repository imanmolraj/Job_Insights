```python
import pandas as pd
```

```python
try:
    df = pd.read_csv('job_descriptions.csv')
except FileNotFoundError:
    print("Error: The file 'job_descriptions.csv' was not found. Please check the file path.")
    exit()
```

```python
df.head()
```

```python
# Filter for work_type = 'Intern'
df_filtered = df[df['Work Type'] == 'Intern']

# Filter for latitude below 10
df_filtered = df_filtered[df_filtered['latitude'] < 10]
```

```python
df_filtered['county_name'] = df_filtered['Country'] 
df_filtered = df_filtered[~df_filtered['county_name'].str.startswith(('A', 'B', 'C', 'D'))]
```

```python
# Filter for job title with no more than 10 characters
df_filtered = df_filtered[df_filtered['Job Title'].str.len() <= 10]

# Filter for company size below 50,000
df_filtered = df_filtered[df_filtered['Company Size'] < 50000]
```

```python
output_file = 'filtered_intern_data.csv'
df_filtered.to_csv(output_file,encoding='utf-8', index=False)

print(f"Filtered data has been successfully saved to {output_file}")
```
